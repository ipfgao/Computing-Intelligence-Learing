```python
import jieba#加载jieba分词模块，没有的可以pip install jieba
```

```python
max_length = 1000000 

sub_file = FILE[:max_length] 
#选取FILE里前1000000赋予subfile
```

```python
#定义分词函数cut，返回list类型
def cut(string):
    return list(jieba.cut(string))
```

```python
TOKENS = cut(sub_file)#对sub_file分词，返回值赋予TOKENS
```

```python
len(TOKENS)
```

```python
from collections import Counter
```

```python
%matplotlib inline
```

```python
words_count = Counter(TOKENS)#统计TOKENS里每个词汇出现的次数
```

```python
print(words_count)
```

```python
words_count.most_common(20)#列出出现最多的20项
```

```python
words_with_fre = [f for w, f in words_count.most_common()]
```
